```python
import numpy as np
from IPython.display import display
import pandas as pd
np.set_printoptions(suppress=True, formatter={'float_kind': '{:0.1f}'.format})
```

```python
df1=pd.read_csv("mobile.csv")
display(df1)
```

```python
df1["First Name"] = df1["Customer Name"].str.split().str[0]
print(df1)
```

```python
df1['company'] = df1['Phone Number'].str[:3].replace({'050': 'Pelephone','051': 'Golan','052': 'Cellcom','053': 'HOT', '054' :'Partner'})
display(df1)
```

```python
display(df1['company'].value_counts())
display(df1['company'].unique())
display(df1['company'].nunique())
```

```python
# def phone_type(Talked,Internet):
#     if Talked>Internet:
#         return "Talker"
#     else:
#         return "Surfer"
# df1['phone_type'] = np.vectorize(phone_type)(df1['Talked'], df1['Internet'])

df1.apply(lambda row: 'Talker' if row['Talked'] > row['Internet'] else 'Surfer', axis=1)

display(df1)
```

```python
sorted_df = df1.sort_values("Talked", ascending=True)
display(sorted_df)
```

```python
sorted_df = df1.sort_values(by=["Talked", "Internet"], ascending=True)
display(sorted_df)
```

```python
display(df1.nlargest(5, "Talked"))
display(df1.nsmallest(3, "Internet"))
```
